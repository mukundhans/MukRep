```python
user_history = pd.read_csv('historical_user_logs.csv')
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
test['is_click'] = -1
data = train.append(test, ignore_index=True)

data['datetime'] =  pd.to_datetime(data['DateTime'])
data['hour'] = data['datetime'].dt.hour
data['minute'] = data['datetime'].dt.minute
data['day'] = data['datetime'].dt.weekday
data['date'] = data['datetime'].dt.day

data['ad_rank'] = data.groupby(['user_id'])['datetime'].rank(method='first').astype(int)
data['user_prod_ad_rank'] = data.groupby(['user_id','product'])['datetime'].rank(method='first').astype(int)
data['user_cam_ad_rank'] = data.groupby(['user_id','campaign_id'])['datetime'].rank(method='first').astype(int)
data['user_prod_cam_ad_rank'] = data.groupby(['user_id','product','campaign_id'])['datetime'].rank(method='first').astype(int)
```

```python
data['lagdate1'] = data.groupby(['user_id'])['datetime'].shift(1)
data['lag_diff1'] = (data['datetime'] - data['lagdate1'])/ np.timedelta64(1, 'D') 
data['lagdate2'] = data.groupby(['user_id'])['datetime'].shift(2)
data['lag_diff2'] = (data['datetime'] - data['lagdate2'])/ np.timedelta64(1, 'D') 
data['lagdate3'] = data.groupby(['user_id'])['datetime'].shift(3)
data['lag_diff3'] = (data['datetime'] - data['lagdate3'])/ np.timedelta64(1, 'D') 
data['lagdate4'] = data.groupby(['user_id'])['datetime'].shift(4)
data['lag_diff4'] = (data['datetime'] - data['lagdate4'])/ np.timedelta64(1, 'D') 
data['lagdate5'] = data.groupby(['user_id'])['datetime'].shift(5)
data['lag_diff5'] = (data['datetime'] - data['lagdate5'])/ np.timedelta64(1, 'D') 

data['prod_lagdate1'] = data.groupby(['user_id','product'])['datetime'].shift(1)
data['prod_lag_diff1'] = (data['datetime'] - data['prod_lagdate1'])/ np.timedelta64(1, 'D') 
data['prod_lagdate2'] = data.groupby(['user_id','product'])['datetime'].shift(2)
data['prod_lag_diff2'] = (data['datetime'] - data['prod_lagdate2'])/ np.timedelta64(1, 'D') 


data['camp_lagdate1'] = data.groupby(['user_id','campaign_id'])['datetime'].shift(1)
data['camp_lag_diff1'] = (data['datetime'] - data['camp_lagdate1'])/ np.timedelta64(1, 'D') 
data['camp_lagdate2'] = data.groupby(['user_id','campaign_id'])['datetime'].shift(2)
data['camp_lag_diff2'] = (data['datetime'] - data['camp_lagdate2'])/ np.timedelta64(1, 'D') 

data['prod_camp_lagdate1'] = data.groupby(['user_id','product','campaign_id'])['datetime'].shift(1)
data['prod_camp_lag_diff1'] = (data['datetime'] - data['prod_camp_lagdate1'])/ np.timedelta64(1, 'D') 
data['prod_camp_lagdate2'] = data.groupby(['user_id','product','campaign_id'])['datetime'].shift(2)
data['prod_camp_lag_diff2'] = (data['datetime'] - data['prod_camp_lagdate2'])/ np.timedelta64(1, 'D') 

data['rank_ratio'] = data.user_prod_ad_rank/data.ad_rank
data['lag_ratio1'] = data.prod_lag_diff1/data.lag_diff1
data['mean_lag'] = (data.lag_diff1+(data.lag_diff2 - data.lag_diff1)+(data.lag_diff3 - data.lag_diff2)
                   +(data.lag_diff4 - data.lag_diff3)+(data.lag_diff5 - data.lag_diff4))/5

train = data[(data['is_click'] != -1)]
test = data[(data['is_click'] == -1)]
```

```python
user_history['interest_flag'] = [ 0 if x == 'view' else 1 for x in user_history['action']]
```

```python
user_history['datetime'] =  pd.to_datetime(user_history['DateTime'])
user_history['date'] = user_history['datetime'].dt.date
user_history['hour'] = user_history['datetime'].dt.hour
user_history['minute'] = user_history['datetime'].dt.minute
user_history['day'] = user_history['datetime'].dt.weekday
```

```python
#User History
user_hist_rate = user_history.groupby("user_id").agg({'interest_flag' : ['size','sum']})\
.reset_index()
user_hist_rate.columns = user_hist_rate.columns.droplevel(0)
user_hist_rate.columns = ['user_id' , 'total_hits' , 'interests']
user_hist_rate['user_int_rate'] = user_hist_rate.interests/user_hist_rate.total_hits

#User Product History
user_prod_hist_rate = user_history.groupby(["user_id" , "product"]).agg({'interest_flag' : ['size','sum']})\
.reset_index()
user_prod_hist_rate.columns = user_prod_hist_rate.columns.droplevel(0)
user_prod_hist_rate.columns = ['user_id' ,'product', 'prod_total_hits' , 'prod_interests']
user_prod_hist_rate['prod_int_rate'] = user_prod_hist_rate.prod_interests/user_prod_hist_rate.prod_total_hits

#Product History
prod_hist_rate = user_history.groupby("product").agg({'interest_flag' : ['size','sum']})\
.reset_index()
prod_hist_rate.columns = prod_hist_rate.columns.droplevel(0)
prod_hist_rate.columns = ['product' , 'ptotal_hits' , 'pinterests']
prod_hist_rate['pint_rate'] = prod_hist_rate.pinterests/prod_hist_rate.ptotal_hits

#hour History
hour_hist_rate = user_history.groupby("hour").agg({'interest_flag' : ['size','sum']})\
.reset_index()
hour_hist_rate.columns = hour_hist_rate.columns.droplevel(0)
hour_hist_rate.columns = ['hour' , 'hour_total_hits' , 'hour_interests']
hour_hist_rate['hour_int_rate'] = hour_hist_rate.hour_interests/hour_hist_rate.hour_total_hits

#day History
day_hist_rate = user_history.groupby("day").agg({'interest_flag' : ['size','sum']})\
.reset_index()
day_hist_rate.columns = day_hist_rate.columns.droplevel(0)
day_hist_rate.columns = ['day' , 'day_total_hits' , 'day_interests']
day_hist_rate['day_int_rate'] = day_hist_rate.day_interests/day_hist_rate.day_total_hits
day_hist_rate

#User day History
user_day_hist_rate = user_history.groupby(["user_id" , "day"]).agg({'interest_flag' : ['size','sum']})\
.reset_index()
user_day_hist_rate.columns = user_day_hist_rate.columns.droplevel(0)
user_day_hist_rate.columns = ['user_id' ,'day', 'user_day_total_hits' , 'user_day_interests']
user_day_hist_rate['user_day_int_rate'] = user_day_hist_rate.user_day_interests/user_day_hist_rate.user_day_total_hits

#prod day History
prod_day_hist_rate = user_history.groupby(["product" , "day"]).agg({'interest_flag' : ['size','sum']})\
.reset_index()
prod_day_hist_rate.columns = prod_day_hist_rate.columns.droplevel(0)
prod_day_hist_rate.columns = ['product' ,'day', 'prod_day_total_hits' , 'prod_day_interests']
prod_day_hist_rate['prod_day_int_rate'] = prod_day_hist_rate.prod_day_interests/prod_day_hist_rate.prod_day_total_hits
prod_day_hist_rate

#User Product Distribution matrix
user_prod_hist_mat = pd.merge(user_prod_hist_rate,user_hist_rate, how='left',on='user_id')
user_prod_hist_mat['prod_dist'] = user_prod_hist_mat.prod_total_hits/user_prod_hist_mat.total_hits
user_prod_hist_mat = user_prod_hist_mat.iloc[:,(0,1,8)]
user_prod_hist_mat = user_prod_hist_mat.pivot(index='user_id' , columns='product',values='prod_dist')
user_prod_hist_mat = pd.DataFrame(user_prod_hist_mat.to_records())

#AOG, Recency, first ad cust
hist_lag_met = user_history.iloc[:,(6,1,2,4)]
user_min_max = hist_lag_met.groupby('user_id').agg({'date' : ['min','max','size']}).reset_index()
user_min_max.columns = user_min_max.columns.droplevel(0)
user_min_max.columns = ['user_id' , 'first_ad' , 'last_ad','freq']
today_date = max(user_min_max['last_ad']) + timedelta(1)
user_min_max['today_date'] = today_date
user_min_max['difference'] = (user_min_max['last_ad'] - user_min_max['first_ad'])/ np.timedelta64(1, 'D') + 1
user_min_max['since_first_ad'] = (user_min_max['today_date'] - user_min_max['first_ad'])/ np.timedelta64(1, 'D') 
user_min_max['since_last_ad'] = (user_min_max['today_date'] - user_min_max['last_ad'])/ np.timedelta64(1, 'D') 
user_min_max['freq'] = user_min_max['freq'].astype('float')
user_min_max['aog'] = user_min_max.difference/user_min_max.freq
user_min_max = user_min_max.iloc[:,(0,3,5,6,7,8)]

#AOG, Recency, first ad cust-prod
user_prod_min_max = hist_lag_met.groupby(['user_id','product']).agg({'date' : ['min','max','size']}).reset_index()
user_prod_min_max.columns = user_prod_min_max.columns.droplevel(0)
user_prod_min_max.columns = ['user_id' , 'product','prod_first_ad' , 'prod_last_ad','prod_freq']
today_date = max(user_prod_min_max['prod_last_ad']) + timedelta(1)
user_prod_min_max['prod_today_date'] = today_date
user_prod_min_max['prod_difference'] = (user_prod_min_max['prod_last_ad'] - 
                                        user_prod_min_max['prod_first_ad'])/ np.timedelta64(1, 'D') + 1
user_prod_min_max['prod_since_first_ad'] = (user_prod_min_max['prod_today_date'] - 
                                            user_prod_min_max['prod_first_ad'])/ np.timedelta64(1, 'D') 
user_prod_min_max['prod_since_last_ad'] = (user_prod_min_max['prod_today_date'] - 
                                           user_prod_min_max['prod_last_ad'])/ np.timedelta64(1, 'D') 
user_prod_min_max['prod_freq'] = user_prod_min_max['prod_freq'].astype('float')
user_prod_min_max['prod_aog'] = user_prod_min_max.prod_difference/user_prod_min_max.prod_freq
user_prod_min_max = user_prod_min_max.iloc[:,(0,1,4,6,7,8,9)]
```

```python
train = pd.merge(train,user_hist_rate, how='left',on='user_id')
train = pd.merge(train,user_prod_hist_rate, how='left',on=['user_id','product'])
train = pd.merge(train,prod_hist_rate, how='left',on='product')
train = pd.merge(train,hour_hist_rate, how='left',on='hour')
train = pd.merge(train,day_hist_rate, how='left',on=['day'])
train = pd.merge(train,user_day_hist_rate, how='left',on=['user_id','day'])
train = pd.merge(train,prod_day_hist_rate, how='left',on=['day','product'])
train = pd.merge(train,user_prod_hist_mat, how='left',on='user_id')
train = pd.merge(train,user_min_max, how='left',on='user_id')
train = pd.merge(train,user_prod_min_max, how='left',on=['user_id','product'])

test = pd.merge(test,user_hist_rate, how='left',on='user_id')
test = pd.merge(test,user_prod_hist_rate, how='left',on=['user_id','product'])
test = pd.merge(test,prod_hist_rate, how='left',on='product')
test = pd.merge(test,hour_hist_rate, how='left',on='hour')
test = pd.merge(test,day_hist_rate, how='left',on=['day'])
test = pd.merge(test,user_day_hist_rate, how='left',on=['user_id','day'])
test = pd.merge(test,prod_day_hist_rate, how='left',on=['day','product'])
test = pd.merge(test,user_prod_hist_mat, how='left',on='user_id')
test = pd.merge(test,user_min_max, how='left',on='user_id')
test = pd.merge(test,user_prod_min_max, how='left',on=['user_id','product'])

#total ads sent to the customers
user_click_rate_train = train.groupby("user_id").agg({'user_id' : ['size']}).reset_index()
user_click_rate_train.columns = user_click_rate_train.columns.droplevel(0)
user_click_rate_train.columns = ['user_id' , 'user_total_ads' ]
user_click_rate_train

train_user_all = train.iloc[:,2:6]
test_user_all = test.iloc[:,2:6]
train_test = train_user_all.append(test_user_all)

user_click_rate_test = train_test.groupby("user_id").agg({'user_id' : ['size']}).reset_index()
user_click_rate_test.columns = user_click_rate_test.columns.droplevel(0)
user_click_rate_test.columns = ['user_id' , 'user_total_ads' ]

train = pd.merge(train,user_click_rate_train, how='left',on='user_id')
test = pd.merge(test,user_click_rate_test, how='left',on='user_id')

train['user_total_ads'] = train.user_total_ads/6
test['user_total_ads'] = test.user_total_ads/8

#user product total ads
user_prod_click_rate_train = train.groupby(["user_id","product"]).agg({'user_id' : ['size']}).reset_index()
user_prod_click_rate_train.columns = user_prod_click_rate_train.columns.droplevel(0)
user_prod_click_rate_train.columns = ['user_id' , 'product','user_prod_total_ads' ]

train_user_all = train.iloc[:,2:6]
test_user_all = test.iloc[:,2:6]
train_test = train_user_all.append(test_user_all)

user_prod_click_rate_test = train_test.groupby(["user_id","product"]).agg({'user_id' : ['size']}).reset_index()
user_prod_click_rate_test.columns = user_prod_click_rate_test.columns.droplevel(0)
user_prod_click_rate_test.columns = ['user_id' ,'product', 'user_prod_total_ads' ]

train = pd.merge(train,user_prod_click_rate_train, how='left',on=['user_id' , 'product'])
test = pd.merge(test,user_prod_click_rate_test, how='left',on=['user_id','product'])

train['user_prod_total_ads'] = train.user_prod_total_ads/6
test['user_prod_total_ads'] = test.user_prod_total_ads/8

#user campaign total ads
user_cam_click_rate_train = train.groupby(["user_id","campaign_id"]).agg({'user_id' : ['size']}).reset_index()
user_cam_click_rate_train.columns = user_cam_click_rate_train.columns.droplevel(0)
user_cam_click_rate_train.columns = ['user_id' , 'campaign_id','user_cam_total_ads' ]

train_user_all = train.iloc[:,2:6]
test_user_all = test.iloc[:,2:6]
train_test = train_user_all.append(test_user_all)

user_cam_click_rate_test = train_test.groupby(["user_id","campaign_id"]).agg({'user_id' : ['size']}).reset_index()
user_cam_click_rate_test.columns = user_cam_click_rate_test.columns.droplevel(0)
user_cam_click_rate_test.columns = ['user_id' , 'campaign_id','user_cam_total_ads' ]

train = pd.merge(train,user_cam_click_rate_train, how='left',on=['user_id' , 'campaign_id'])
test = pd.merge(test,user_cam_click_rate_test, how='left',on=['user_id','campaign_id'])

train['user_cam_total_ads'] = train.user_cam_total_ads/6
test['user_cam_total_ads'] = test.user_cam_total_ads/8


#user webpage_id total ads
user_web_click_rate_train = train.groupby(["user_id","webpage_id"]).agg({'user_id' : ['size']}).reset_index()
user_web_click_rate_train.columns = user_web_click_rate_train.columns.droplevel(0)
user_web_click_rate_train.columns = ['user_id' , 'webpage_id','user_web_total_ads' ]

train_user_all = train.iloc[:,2:6]
test_user_all = test.iloc[:,2:6]
train_test = train_user_all.append(test_user_all)

user_web_click_rate_test = train_test.groupby(["user_id","webpage_id"]).agg({'user_id' : ['size']}).reset_index()
user_web_click_rate_test.columns = user_web_click_rate_test.columns.droplevel(0)
user_web_click_rate_test.columns = ['user_id' , 'webpage_id','user_web_total_ads' ]

train = pd.merge(train,user_web_click_rate_train, how='left',on=['user_id' , 'webpage_id'])
test = pd.merge(test,user_web_click_rate_test, how='left',on=['user_id','webpage_id'])

train['user_web_total_ads'] = train.user_web_total_ads/6
test['user_web_total_ads'] = test.user_web_total_ads/8

#Prod click rate
prod_click_rate = train.groupby("product").agg({'is_click' : ['size' , 'sum']}).reset_index()
prod_click_rate.columns = prod_click_rate.columns.droplevel(0)
prod_click_rate.columns = ['product' , 'prod_total_ads' , 'prod_total_clicks']
prod_click_rate['prod_click_rate'] = prod_click_rate.prod_total_clicks/prod_click_rate.prod_total_ads

train = pd.merge(train,prod_click_rate, how='left',on='product')
test = pd.merge(test,prod_click_rate, how='left',on='product')

hour_click_rate = train.groupby("hour").agg({'is_click' : ['size' , 'sum']}).reset_index()
hour_click_rate.columns = hour_click_rate.columns.droplevel(0)
hour_click_rate.columns = ['hour' , 'hour_total_ads' , 'hour_total_clicks']
hour_click_rate['hour_click_rate'] = hour_click_rate.hour_total_clicks/hour_click_rate.hour_total_ads

train = pd.merge(train,hour_click_rate, how='left',on='hour')
test = pd.merge(test,hour_click_rate, how='left',on='hour')

cam_click_rate = train.groupby("campaign_id").agg({'is_click' : ['size' , 'sum']}).reset_index()
cam_click_rate.columns = cam_click_rate.columns.droplevel(0)
cam_click_rate.columns = ['campaign_id' , 'cam_total_ads' , 'cam_total_clicks']
cam_click_rate['cam_click_rate'] = cam_click_rate.cam_total_clicks/cam_click_rate.cam_total_ads
cam_click_rate

train = pd.merge(train,cam_click_rate, how='left',on='campaign_id')
test = pd.merge(test,cam_click_rate, how='left',on='campaign_id')

web_click_rate = train.groupby("webpage_id").agg({'is_click' : ['size' , 'sum']}).reset_index()
web_click_rate.columns = web_click_rate.columns.droplevel(0)
web_click_rate.columns = ['webpage_id' , 'web_total_ads' , 'web_total_clicks']
web_click_rate['web_click_rate'] = web_click_rate.web_total_clicks/web_click_rate.web_total_ads
web_click_rate

train = pd.merge(train,web_click_rate, how='left',on='webpage_id')
test = pd.merge(test,web_click_rate, how='left',on='webpage_id')

# Product Cam
prod_cam_cr = train.groupby(["product" , "campaign_id"]).agg({'is_click' : ['size','sum']}).reset_index()
prod_cam_cr.columns = prod_cam_cr.columns.droplevel(0)
prod_cam_cr.columns = ['product' ,'campaign_id', 'pc_hits' , 'pc_clicks']
prod_cam_cr['pc_click_rate'] = prod_cam_cr.pc_clicks/prod_cam_cr.pc_hits

train = pd.merge(train,prod_cam_cr, how='left',on=['product','campaign_id'])
test = pd.merge(test,prod_cam_cr, how='left',on=['product','campaign_id'])

# prod web
web_cam_cr = train.groupby(["product" , "webpage_id"]).agg({'is_click' : ['size','sum']}).reset_index()
web_cam_cr.columns = web_cam_cr.columns.droplevel(0)
web_cam_cr.columns = ['product' ,'webpage_id', 'wc_hits' , 'wc_clicks']
web_cam_cr['wc_click_rate'] = web_cam_cr.wc_clicks/web_cam_cr.wc_hits

train = pd.merge(train,web_cam_cr, how='left',on=['product','webpage_id'])
test = pd.merge(test,web_cam_cr, how='left',on=['product','webpage_id'])

# cam web
cam_web_cr = train.groupby(["campaign_id" , "webpage_id"]).agg({'is_click' : ['size','sum']}).reset_index()
cam_web_cr.columns = cam_web_cr.columns.droplevel(0)
cam_web_cr.columns = ['campaign_id' ,'webpage_id', 'cw_hits' , 'cw_clicks']
cam_web_cr['cw_click_rate'] = cam_web_cr.cw_clicks/cam_web_cr.cw_hits

train = pd.merge(train,cam_web_cr, how='left',on=['campaign_id','webpage_id'])
test = pd.merge(test,cam_web_cr, how='left',on=['campaign_id','webpage_id'])

# Product hour
prod_hr_cr = train.groupby(["product" , "hour"]).agg({'is_click' : ['size','sum']}).reset_index()
prod_hr_cr.columns = prod_hr_cr.columns.droplevel(0)
prod_hr_cr.columns = ['product' ,'hour', 'ph_hits' , 'ph_clicks']
prod_hr_cr['ph_click_rate'] = prod_hr_cr.ph_clicks/prod_hr_cr.ph_hits

train = pd.merge(train,prod_hr_cr, how='left',on=['product','hour'])
test = pd.merge(test,prod_hr_cr, how='left',on=['product','hour'])

# hour web
web_hour_cr = train.groupby(["hour" , "webpage_id"]).agg({'is_click' : ['size','sum']}).reset_index()
web_hour_cr.columns = web_hour_cr.columns.droplevel(0)
web_hour_cr.columns = ['hour' ,'webpage_id', 'wh_hits' , 'wh_clicks']
web_hour_cr['wh_click_rate'] = web_hour_cr.wh_clicks/web_hour_cr.wh_hits

train = pd.merge(train,web_hour_cr, how='left',on=['hour','webpage_id'])
test = pd.merge(test,web_hour_cr, how='left',on=['hour','webpage_id'])

# cam hour
cam_hour_cr = train.groupby(["campaign_id" , "hour"]).agg({'is_click' : ['size','sum']}).reset_index()
cam_hour_cr.columns = cam_hour_cr.columns.droplevel(0)
cam_hour_cr.columns = ['campaign_id' ,'hour', 'ch_hits' , 'ch_clicks']
cam_hour_cr['ch_click_rate'] = cam_hour_cr.ch_clicks/cam_hour_cr.ch_hits

train = pd.merge(train,cam_hour_cr, how='left',on=['campaign_id','hour'])
test = pd.merge(test,cam_hour_cr, how='left',on=['campaign_id','hour'])

#product cat1
pc1_click_rate = train.groupby("product_category_1").agg({'is_click' : ['size' , 'sum']}).reset_index()
pc1_click_rate.columns = pc1_click_rate.columns.droplevel(0)
pc1_click_rate.columns = ['product_category_1' , 'pc1_total_ads' , 'pc1_total_clicks']
pc1_click_rate['pc1_click_rate'] = pc1_click_rate.pc1_total_clicks/pc1_click_rate.pc1_total_ads
pc1_click_rate

train = pd.merge(train,pc1_click_rate, how='left',on='product_category_1')
test = pd.merge(test,pc1_click_rate, how='left',on='product_category_1')

#user_group_id
ugid_click_rate = train.groupby("user_group_id").agg({'is_click' : ['size' , 'sum']}).reset_index()
ugid_click_rate.columns = ugid_click_rate.columns.droplevel(0)
ugid_click_rate.columns = ['user_group_id' , 'ugid_total_ads' , 'ugid_total_clicks']
ugid_click_rate['ugid_click_rate'] = ugid_click_rate.ugid_total_clicks/ugid_click_rate.ugid_total_ads

train = pd.merge(train,ugid_click_rate, how='left',on='user_group_id')
test = pd.merge(test,ugid_click_rate, how='left',on='user_group_id')

#age level
al_click_rate = train.groupby("age_level").agg({'is_click' : ['size' , 'sum']}).reset_index()
al_click_rate.columns = al_click_rate.columns.droplevel(0)
al_click_rate.columns = ['age_level' , 'al_total_ads' , 'al_total_clicks']
al_click_rate['al_click_rate'] = al_click_rate.al_total_clicks/al_click_rate.al_total_ads

train = pd.merge(train,al_click_rate, how='left',on='age_level')
test = pd.merge(test,al_click_rate, how='left',on='age_level')

#User Depth 
ud_click_rate = train.groupby("user_depth").agg({'is_click' : ['size' , 'sum']}).reset_index()
ud_click_rate.columns = ud_click_rate.columns.droplevel(0)
ud_click_rate.columns = ['user_depth' , 'ud_total_ads' , 'ud_total_clicks']
ud_click_rate['ud_click_rate'] = ud_click_rate.ud_total_clicks/ud_click_rate.ud_total_ads

train = pd.merge(train,ud_click_rate, how='left',on='user_depth')
test = pd.merge(test,ud_click_rate, how='left',on='user_depth')

#day
wday_click_rate = train.groupby("day").agg({'is_click' : ['size' , 'sum']}).reset_index()
wday_click_rate.columns = wday_click_rate.columns.droplevel(0)
wday_click_rate.columns = ['day' , 'wday_total_ads' , 'wday_total_clicks']
wday_click_rate['wday_click_rate'] = wday_click_rate.wday_total_clicks/wday_click_rate.wday_total_ads

train = pd.merge(train,wday_click_rate, how='left',on='day')
test = pd.merge(test,wday_click_rate, how='left',on='day')

# day prod
dp_hour_cr = train.groupby(["product" , "day"]).agg({'is_click' : ['size','sum']}).reset_index()
dp_hour_cr.columns = dp_hour_cr.columns.droplevel(0)
dp_hour_cr.columns = ['product' ,'day', 'dp_hits' , 'dp_clicks']
dp_hour_cr['dp_click_rate'] = dp_hour_cr.dp_clicks/dp_hour_cr.dp_hits

train = pd.merge(train,dp_hour_cr, how='left',on=['product','day'])
test = pd.merge(test,dp_hour_cr, how='left',on=['product','day'])
```
